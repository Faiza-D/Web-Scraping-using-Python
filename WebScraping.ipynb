```python
# Find the table by class name or id
table = soup.find('table', {'id': 'example-table'})
# Extract rows from the table
rows = table.find_all('tr')
for row in rows:
    cells = row.find_all('td')
    data = [cell.text for cell in cells]
    print(data)  # Prints each row's data
```

```python
# Extract all links on the page
links = soup.select('a[href]')
for link in links:
    print(link['href'])
```

Best Practices for Scraping with Selenium

-Use headless mode for faster scraping.
-Respect rate limits and add delays between requests.
-Randomize your actions to mimic human behavior.
-Check robots.txt of the site to ensure you follow their scraping policy.
-Save progress: Write scraped data to CSV, JSON, or databases in intervals.
